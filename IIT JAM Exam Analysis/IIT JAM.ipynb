```python
def automate_keys(year,df_year,mcq_key,msq_key,range_key):
    df[year] =np.zeros(60)
    
    df_year[msq_key] = df_year[msq_key].str.replace(';',',')
    
    df.loc[:29, year] = df_year[mcq_key].values
    df.loc[30:39, year] = df_year.loc[:9, msq_key].values
    df.loc[40:, year] = df_year.loc[:19, range_key].str.split(':',expand=True)[0].values
    
    df[year] = df[year].str.strip()
```
